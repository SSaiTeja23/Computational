```python
from geopy.geocoders import Nominatim
import pycountry
def country_check(pos):            #we have defined a function using pycountry and geocoders lib for linking city name with country
    geolocator = Nominatim()
    location = geolocator.geocode(pos)
    loc=str(location.address)
    for country in pycountry.countries:
        if country.name in loc:
            return country.name
```

```python
cities=[]
countries=[]
cities=data['City'].unique().tolist()
for i in cities:
    try:
        countries.append(country_check(i))  #Calling the defined function
    except:
        countries.append('Country Not recognized')   #Some of the cities might not be detected by library
                                                     #Lets define them as not recognized for now
```

```python
import pycountry_convert as pc
continents=[]
for j in countries:
    try:         #Now link the recognized country with the continent
        country_code = pc.country_name_to_country_alpha2(j, cn_name_format="default")
        continent_name = pc.country_alpha2_to_continent_code(country_code)
        continents.append(continent_name)
    except:
        continents.append('Continent not recognized')
```

```python
dict_1={}
dict_2={}
dict_1=dict(zip(cities,countries))
#dict_1['London']='England'   #For unrecognized cities define there country manually
dict_1['can & sun']='Australia'
dict_1['Jerusalem']='Israel'
dict_1['Ulaanbaatar']='Mongolia'
dict_1['Karachi']='Pakistan'
dict_1['peshawar']='Pakistan'
dict_1['rawalpindi']='Pakistan'
dict_1['Lahore']='Pakistan'
dict_1['Seoul']='Sri lanka'
dict_1['Busan']='Sri lanka'
dict_1['Daegu']='Sri lanka'
dict_1['Aleppo']='Syria'
dict_1['Taipei']='Taiwan'
dict_1['Chiang mai']='Thailand'
dict_1['Pattaya City']='Thailand'
dict_1['Phuket']='Thailand'
dict_1['Ho Chi Minh City']='United Arab Emirates'
dict_2=dict(zip(countries,continents))
```

```python
data['Country']=data['City'].map(dict_1)
data['Continent']=data['Country'].map(dict_2)
#data=data[['tweet_count','City','Country','Continent','tweet_text','date','hashtags','link']]
data
```
