```python
import pandas as pd
df= pd.read_csv(r"C:\Users\deepa\Desktop\Computational_Methods_Project\Unnecessary\sample_output_4.csv")
```

```python
from langdetect import detect
lang=[]
for i in df['tweet_text']:
    try:
        lang.append(detect(i))   #Here we Have detected language and appended to list
    except:
        lang.append("Lang not detected")  #Language not detected
        continue
dfObj= pd.DataFrame(lang, columns = ['Language'])  
dfObj.to_csv(r"C:\Users\deepa\Desktop\Computational_Methods_Project\Unnecessary\lang_new_1.csv")
```
